### Grab YOLO model
Grab it and show the details of the model

```python
from ultralytics import YOLO
model = YOLO("yolo11l.pt")
model.info()
```

### Show image to analyze

```python
from IPython.display import Image
Image(filename='image.jpg')
```

### Analyze image
Rename the file name to much your image.

```python
results = model.predict("image.jpg", save=True)
result = results[0]
result
```

### Show analyzed image
The resulting image is saved in 'runs/detect/predict'

```python
from IPython.display import Image
Image(filename='runs/detect/predict/image.jpg')
```

The details of the detect objects are in "boxes". Let's see how many objects have been detected. It doesn't include just "person". The COCO dataset has 80 labels in total

```python
len(result.boxes)
```

Examine the details of each box

```python
for box in result.boxes:
  class_id = result.names[box.cls[0].item()]
  cords = box.xyxy[0].tolist()
  cords = [round(x) for x in cords]
  conf = round(box.conf[0].item(), 2)
  print("Object type:", class_id)
  print("Coordinates:", cords)
  print("Probability:", conf)
  print("---")
```

### Count the amount of people in the image
Person is number "0" in the COCO dataset

```python
num_attendees = 0
for box in result.boxes:
  #print(int(box.cls[0].item()))
  if int(box.cls[0].item()) == 0:
    num_attendees += 1
```

```python
from IPython.display import Markdown, display
display(Markdown("# Number of people: " + str(num_attendees)))
```
